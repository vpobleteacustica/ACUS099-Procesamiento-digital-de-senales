### Variable independiente en la representación matemática de una señal.

* Las señales son representadas matemáticamente como una **función**. Estas pueden ser funciones de **una variable independiente**, o **más de una variable independiente**.

$$
\begin{align}
f_{1}(t) &= t^{2}\\
f_{2}(x,y) &= x^{2}+2xy+x^{2}
\end{align}
$$


* Es común en procesamiento de señales referirse a la variable independiente en la representación matemática de una señal como el **tiempo**. 
* La variable tiempo puede ser **continua** o **discreta**.
* Las señales de **tiempo continuo** se definen a lo largo de un continuo de tiempo y son representadas por una variable **independiente continua**.
* Las señales de tiempo continuo son llamadas como **señales análogas**.

+ Las señales de **tiempo discreto** son definidas en **tiempos discretos**. Por lo que la variable independiente toma **valores discretos**. Así, las señales de tiempo discreto son representadas como **secuencias de números**.

+ Además de ser la variable independiente, ya sea continua o discreta, la **amplitud** de la señal puede ser **continua** o **discreta**.

+ Las señales para las cuales **tanto el tiempo como la amplitud son discretos** son llamadas **señales digitales**.

### Señales en tiempo discreto

> + Cuando muestreamos (sampleamos) una señal de tiempo continuo a una frecuencia de muestreo $f_{s}$, convertimos la señal en una señal de tiempo discreto.
> + Cuán frecuentemente necesitamos muestrear, está gobernado por el **Teorema del muestreo**. 

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

f = 50
t = np.linspace(0, 0.05, 1000)

plt.figure(figsize=(30, 4))
plt.plot(t, 10 * np.sin(2*np.pi*f*t))
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title(r'$x(t),\,\, t \geq 0 $',fontsize= 30)
plt.xlim([0, 0.05])

plt.show()
```

```python
f = 50
n = np.arange(50);
dt = 0.07/f
x = np.sin(2 * np.pi * 50 * n * dt)

plt.figure(figsize=(30, 4))
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title(r'$x[n],\,\, n \geq 0 $',fontsize= 30)
plt.stem(n, x);
```

### Ploteando señales de valor complejo

> + Consideremos la señal de tiempo continuo de valor complejo: 
$$
\begin{align}
x(t)&=e^{\,j100 \pi t}
\end{align}
$$

> + Plotear esta señal como la parte real y la parte imaginaria de la señal.
> + Plotear la magnitud y la fase.

```python
t = np.linspace(0, 0.05, 1000)

plt.figure(figsize=(30, 4))
plt.plot(t, np.exp(2j*np.pi*50*t).real )
plt.xlabel('t')
plt.ylabel('Re x(t)')
plt.title(r'Parte real de $x(t)=e^{j 100 \pi t}$',fontsize= 30)
plt.xlim([0, 0.05])

plt.figure(figsize=(30, 4))
plt.plot(t, np.exp(2j*np.pi*50*t).imag);
plt.xlabel('t');
plt.ylabel('Im x(t)');
plt.title(r'Parte imaginaria de $x(t)=e^{j 100\pi t}$',fontsize= 30);
plt.xlim([0, 0.05]);

plt.show()
```

```python
t = np.linspace(0, 0.05, 1000)

plt.figure(figsize=(30, 4))
plt.plot(t, np.abs(np.exp(2j*np.pi*50*t)) )
plt.xlabel(r'\$t\$')
plt.ylabel(r'\$|x(t)|\$')
plt.title(r'Magnitud de $x(t)=e^{j 100 \pi t}$',fontsize= 30)
plt.xlim([0, 0.05])

plt.figure(figsize=(30, 4))
plt.plot(t, np.angle(np.exp(2j*np.pi*50*t))*360/(2*np.pi))
plt.xlabel('\$t\$')
plt.ylabel(r'\$\angle x(t)\$')
plt.title(r'Fase de $x(t)=e^{j 100 \pi t}$',fontsize= 30)
plt.xlim([0, 0.05])

plt.show()
```

### Señal delta unitario (delta de Kronecker)

Esta señal denotada como $\delta[n]$ no es una función en el sentido convencional. Esta se define como:

\begin{align}
\delta[n] &= \begin{cases}
1 & \text{for } n = 0 \\
0 & \text{otherwise}
\end{cases}
\end{align}

Esta función es también conocida como [*delta de Kronecker*](https://en.wikipedia.org/wiki/Kronecker_delta). La señal es un **impulso unitario**. Sus propiedades más importantes son:

1. **Propiedad de Sifting**

    \begin{align}
    \sum_{k=-\infty}^{\infty} x[k] \cdot \delta[n - k] &= x[n]
    \end{align}
    
    Esta importante propiedad implica: $\sum_{k = -\infty}^{\infty} x[k] = 1$.
    
2. **Propiedad de Linealidad**
    \begin{align}
    a \cdot \delta[n] + b \cdot \delta[n] = (a+b) \cdot \delta[n]
    \end{align}

Esta secuencia impulso unitario de Dirac tiene relevancia tanto en análisis de señales como también en el estudio de sistemas. Se usa fundamentalmente en el análisis de sistemas discretos **lineales** e **invariantes en el tiempo**, a través de las **respuestas impulso** de aquellos sistemas.

```python
import numpy as np
import matplotlib.pyplot as plt

#Defining the delta function
def delta(n):
    if n == 0:
        return 1
    else:
        return 0
   
#Defining lists

h_ = []
n = 7

#Writing the h[n] function in terms of delta function
for i in range(-n,n+1):
    h = delta(i)
    h_.append(h)

plt.figure(figsize=(30, 4))
plt.stem(range(-n,n+1),h_, '--')
plt.ylim([-1,1])
plt.xlim([-n,n])
plt.xlabel('$n$')
plt.ylabel('$h[n]$')
plt.title('$h[n] = \delta [n]$',fontsize= 30)
plt.grid(True)    
```
