### Transformada z:

> + Es una transformada más general que la transformada de Fourier.

> + Se usa para obtener estructuras filtros digitales.

> + Generalmente, se usa la transformada z unilateral, es decir:

$$
\begin{align*}
X(z) & = \sum_{n=0}^{\infty}x[n]z^{-n}
\end{align*}
$$

> + donde $z$ es una variable compleja definida como:

$$
\begin{align*}
z  & = r\cdot e^{j\omega} 
\end{align*}
$$

> + La transformada z se interpreta usando el plano-z.

> + El cŕculo unitario en el plano z, se define como: 

$$
\begin{align*}
|z|  & = 1 
\end{align*}
$$

> + En este plazo-z, $\omega$ es el **ángulo** entre el vector que va desde el origen a un punto $z$ sobre el círculo unitario, y el eje real del plano z.

> + Si se observa el círculo unitario y se evalúa la transformada z sobre este círculo, se vé que corresponde a la transformada de Fourier!
    
$$
\begin{align*}
X(z) & = X(r\cdot e^{j\omega})\\
     & = X(r\cdot e^{j\omega})
     & = X(e^{j\omega}) \\
     & = \sum_{n=0}^{\infty}x[n]\left(e^{j\omega}\right)^{-n}\\
     & = \sum_{n=0}^{\infty}x[n]e^{-j\omega n}
\end{align*}
$$
    
> + Si evaluamos $X(z)$ en puntos sobre el círculo unitario, empezando en $\omega = 0$ es decir, $z=1$, pasando por $\omega = \pi/2$ esto es, $z=j$ hasta $\omega = \pi$, es decir, $z=-1$, entonces vamos a obtener la transformada de Fourier para $0\leq \omega \leq \pi$.
    
> + Así podemos continuar moviéndonos sobre el círculo unitario y obtener la transformada de Fourier desde $\pi\leq \omega \leq 2\pi$.
    
> + Ya que la transformada de Fourier la representamos gráficamente usando un eje de frecuencia lineal, ahora, sobre el círculo unitario en el plano-z, se interpreta la frecuencia como si se doblara ese eje de frecuencia lineal por encima del circulo unitario, con $\omega = 0$ en $z=1$ y $\omega = \pi$ en $z=-1$.
    
> + Con esta interpretación, se captura la **periodicidad** natural en frecuencia de la trnaformada de Fourier, ya que un cambio del ángulo $\omega$ de $2\pi$ radianes en el plano-z, corresponde a atravezar una vez el círculo unitario y regresar al mismo punto exactamente.

### Región de convergencia:

> + De la definición se puede dar que una suma infinita no se siempre **finita** y por lo tanto, puede que la sumatoria sea **no convergente** y la transformada z en consecuencia, **no exista**.

> + Se llama región de convergencia (**ROC**), aquellos valores de z para los que converge la transformada z.
